```python
np.random.seed(42)

# Colors
bk_color = (0, 0, 45/255)
line_color1 = "#d7ffff"
# line_color2 = "#ff87ff"

# Canvas
fig, ax = plt.subplots(figsize=(8,8))
ax.set_facecolor(bk_color)

# Line Equations
x_min = -10
x_max = 10
y_min = -4.5
y_max = 5
sigma = 2
x = np.arange(x_min, x_max, .1)

def g(x, y, sigma):
    r = np.sqrt(x**2 + y**2)
    return np.exp(-r**2/(2*sigma**2))

# Lines
for h in np.arange(y_min, y_max, .5):
    fuzz = np.random.normal(0,0.2)
    y1 = np.repeat(h, len(x))
    amp = g(x, y1, sigma)
    y = h + amp*np.sin(5*x)
    ax.plot(x+fuzz, y, color=line_color1, linewidth=1)
# ax.plot(x1, y2, color=line_color1, linewidth=1)
# ax.plot(x1, y3, color=line_color1, linewidth=1)
# ax.plot(x, y2, color=line_color2, linewidth=30)

# Axes
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

# Save
# ax.axes.xaxis.set_visible(False)
# ax.axes.yaxis.set_visible(False)
# plt.savefig("img/waves1.png")

plt.show()
```

```python
np.random.seed(42)

# Colors
bk_color = (0, 0, 75/255)
# line_color1 = "#d7ffff"
# line_color2 = "#ff87ff"

# Canvas
fig, ax = plt.subplots(figsize=(8,8))
ax.set_facecolor(bk_color)

# Line Equations
n_lines = 100
x_min = -10
x_max = 10
y_min = -4.5
y_max = 5
a = 3
sigma = 1
omega = 9
x = np.arange(x_min, x_max, .1)

def g(x, y, sigma, a):
    r = np.sqrt(x**2 + y**2)
    return a*np.exp(-r**2/(2*sigma**2))

# Lines
for i in np.arange(n_lines):
    fuzz_x = np.random.normal(0,0.5)
    h = np.random.normal(0,.5)
    y1 = np.repeat(h, len(x))
    amp = g(x, y1, sigma, a)
    y = h + amp*np.sin(omega*x)
    fuzz_c = np.random.uniform(0,.2)
    line_color = (np.clip(0.7+fuzz_c,0,1), 0.9, 1.0)
    ax.plot(x+fuzz_x, y, color=line_color, linewidth=1.0)
# Axes
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

# Save
# ax.axes.xaxis.set_visible(False)
# ax.axes.yaxis.set_visible(False)
# plt.savefig("img/waves1.png")

plt.show()
```
